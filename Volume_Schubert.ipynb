```julia
#computes volume of a Schubert matroid polytope
using Random: randperm
using SmallCollections: SmallBitSet, Combinatorics

const S = SmallBitSet{UInt32}

N = 5
function height(a)
    h = 0
    s = 0
    for i in 1:N
        k = SmallBitSet{UInt32}([i])
        
        if k < a
            s = s+1
        end
        h = h +s
    end
    return h, s
end
a = SmallBitSet{UInt32}([1,2,5])
height(a) 

function is_less(a, b)
    s1 = 0
    s2 = 0
    for i in 1:N
        k = SmallBitSet{UInt32}([i])
        if k < a
           s1 = s1+1
        end
        if k < b 
            s2 = s2+1
        end
        if s1 > s2 
            return false
        end
    end
    return true
end

h = Combinatorics.subsets(7)
```

```julia
function delta(subset)
    delta = 0
    for s in collect(Combinatorics.permutations(N+1))
        print
        val = true
        for i in 1:N
            k = SmallBitSet{UInt32}([i])
            #that means there is a one
            if k < subset
                if !(s[i]>s[i+1])
                    val = false
                    break
                end
            end
            if !(k < subset)
                if !(s[i+1]>s[i])
                    val = false
                    break
                end
            end

        end
        if val
            delta = delta+1
        end
    end
    return delta
end

# sums up delta i.e. computes volume of Schubert, with 1 at entry 1, 0 at entry n and free extension in the positions after elements in the subset. For example [1,3], N = 5 corresponds to the Schubert matroid 1,1,0,1,0,0

function delta_leq(a)
    sum =0
    for b in Combinatorics.subsets(N)
        if is_less(b, a) && height(a)[2] == height(b)[2]
            sum = sum + delta(b)
        end
    end
    return sum
end

println(delta_leq(SmallBitSet{UInt16}([2,3,5])))
println(delta_leq(SmallBitSet{UInt16}([1,4,5])))


```
